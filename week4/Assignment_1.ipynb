# Assignment 1: GroupBy 

```python
import pandas as pd
```

### Statistics on Occupations. 

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
#users.head()
users
```

### Get the mean age per occupation

```python
users.groupby(by='occupation').mean()
```

### Get the Male ratio per occupation and sort it from the most to the least

```python
# hint: use function to recode gender as numarical value
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
```

```python
for i in users.index:
    users.loc[i,'gender_num'] = gender_to_numeric(users.loc[i,'gender'])
    
users[users['gender'] == 'M'].groupby(by=['occupation'])['gender_num'].value_counts().sort_values(ascending=False)
```

### For each occupation, calculate the minimum and maximum ages

```python
users.groupby(by='occupation').agg({'age': ['min', 'max']})
```

### For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by=['occupation','gender'])['age'].mean()
```

### [dificult] For each occupation present the percentage of women and men

```python
pd.crosstab(users['occupation'],users['gender']).apply(lambda r: r/r.sum(), axis=1)
```
